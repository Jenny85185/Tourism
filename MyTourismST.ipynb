```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Tourism1.xlsx')
df
```

```python
excursion_counts = df['Type_of_Excursion'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(8, 4))
sns.barplot(x=excursion_counts.index, y=excursion_counts.values)

# Adding titles and labels
plt.title('Number of Excursions by Type')
plt.xlabel('Type of Excursion')
plt.ylabel('Count')

# Display the plot
plt.xticks(rotation=45)  # Rotate the x-axis labels if they are long
plt.show()
```

```python
# Grouping data by season and counting the number of excursions
season_counts = df['Season'].value_counts().sort_index()  # Sorting by season (if needed)

# Plotting a bar chart
plt.figure(figsize=(8, 4))
sns.barplot(x=season_counts.index, y=season_counts.values, color= "skyblue")

# Adding title and labels
plt.title('Number of Excursions by Season')
plt.xlabel('Season')
plt.ylabel('Number of Excursions')

# Rotating x-axis labels if needed
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

```python
# Grouping data by season and type of excursion, counting the number of excursions
season_excursion_counts = df.groupby(['Season', 'Type_of_Excursion']).size().reset_index(name='Count')

# Plotting the graph
plt.figure(figsize=(8, 4))

# Plotting a line chart where each type of excursion will have its own line
sns.lineplot(x='Season', y='Count', hue='Type_of_Excursion', data=season_excursion_counts, marker='o', markersize=8)

# Adding title and labels
plt.title('Number of Excursions by Season and Type')
plt.xlabel('Season')
plt.ylabel('Number of Excursions')

# Rotating x-axis labels if needed
plt.xticks(rotation=45)

# Display the plot
plt.grid(True)
plt.show()


```
